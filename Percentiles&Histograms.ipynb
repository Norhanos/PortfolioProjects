```python
""""


This cell is used to create a function I call, "PercentilesAndHistograms".
The purpose of this function is to study the distribution of the relationship 
between two parameters of interest. This distribution is studied by computing 
percentile curves for the relationship and then dividing the x-axis parameter 
into slices and creating an overlapping histogram plot to showcase the 
distribution of points in each one of those slices.

This function takes the following arguments: 1.) a
parameter to place on the x-axis in the form of a numpy array 2.) a parameter 
to place on the y-axis in the form of a numpy array 3.) a slice width in the 
x-axis parameter to bin by 4.) the minimum value of the x-axis parameter that you 
would like to use 5.) the maximum value of the x-axis parameter that you would
like to use 6.) the number of bins you would like to use for your histograms 
7.) a string which includes the x-axis label and 8.) a string which includes 
the y-axis label. 

The following is the output of the funciton:

1.) Scatter Plot: A scatter plot with the x-axis parameter on the x-axis and the 
y-axis parameter on the y-axis. This plot includes percentile curves for the 
25th, 50th, and 75th percentiles. These percentiles are computed for each slice
in the x-axis parameter. The slices are separated by different colors.
The corner of this plot includes the correlation coefficient (CC) of the 
relationship and the total number of points shown in the plot. 

2.) Histogram: This is a plot of overlapping histograms where each histogram
represents all of the points in one of the slices. For example, if you have
five slices (five colors in the scatter plot above) then you will have five
overlapping histograms. The corner legend indicates the slice limits associated
with each color and the number of points used in each slice.

3.) log Histogram: This is the same as the histogram above, except now with
a log scale used on the y-axis.

4.) Rank-Ordered log Histogram: This is the same as the histogram above, except
now each histogram is rank-ordered so that its peak is always at 1. 

5.) Shifted Rank-Ordered log Histogram: This is the same as the histogram above,
except now each histogram is shifted in the x-axis direction so that all of 
the histograms peak at the same x-axis value. 




"""

#add a shifted plot
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import pylab

def PercentilesAndHistograms(xparameter,yparameter,binwidth,binlimit1,binlimit2,npbins,xlabel,ylabel):

    numofbins= round ( (binlimit2-binlimit1) / binwidth )
    binedges=[]
    for i in range(numofbins+1):
        binedges.append(binlimit1 + (binwidth*i))
    binedges=np.asarray(binedges) 


    colors=plt.cm.get_cmap('jet', numofbins)
    
    plt.figure (figsize = (20, 10))
    labelsize = 40
    ticksize = 35
    plt.xticks(size = ticksize)
    plt.yticks(size = ticksize)


    plt.scatter(xparameter,yparameter,s=1,color='gray')
    for i in range(len(binedges)-1):
        thebin=np.where((xparameter>binedges[i]) & (xparameter<binedges[i+1]))[0]
        p25=np.percentile(yparameter[thebin],25)
        p50=np.percentile(yparameter[thebin],50)
        p75=np.percentile(yparameter[thebin],75)
        plt.plot([binedges[i],binedges[i+1]],[p25,p25],linewidth=5,linestyle=":",color=colors(i))
        plt.plot([binedges[i],binedges[i+1]],[p50,p50],linewidth=5,linestyle="--",color=colors(i))
        plt.plot([binedges[i],binedges[i+1]],[p75,p75],linewidth=5,color=colors(i))
    
 
    cc=np.corrcoef(xparameter,yparameter)[0][1]
    lss = [None,None,None, '--', ":"]
    lw= [0,0,3,3,3]
    lines = [Line2D([0], [0], color='black', linewidth=lw[l], linestyle=lss[l]) for l in range(len(lss))]
    labels = [" ".join(["CC=",str(round(cc,3))])," ".join(["points=",str(len(xparameter))]),'75%', '50%', '25%']
    plt.legend(lines,labels,fontsize=15)
    plt.title("Scatter Plot with Percentiles",size=25)    
    plt.xlabel(xlabel,size=20)
    plt.ylabel(ylabel,size=20)
    plt.grid("True")
    plt.xlim(binlimit1,binlimit2)
    
    plt.figure (figsize = (20, 10))
    labelsize = 40
    ticksize = 35
    plt.xticks(size = ticksize)
    plt.yticks(size = ticksize)

    for i in range(len(binedges)-1):

        thebin=np.where((xparameter>binedges[i]) & (xparameter<binedges[i+1]))[0]
        freq,vals=np.histogram(yparameter[thebin],bins=npbins)
        bincenters = [(vals[q]+vals[q+1])/2 for q in range(len(vals)-1)]
        plt.step(bincenters,freq,linewidth=3,color=colors(i),label=" ".join([str(round(binedges[i],2)),"<",xlabel,"<",str(round(binedges[i+1],2)),";points=",str(len(thebin))]))

    plt.xlabel(ylabel,size=20)
    pylab.legend(loc='upper left',fontsize=15)
    plt.grid("True")
    plt.title(r"Histogram",size=25)
    
    plt.figure (figsize = (20, 10))
    labelsize = 40
    ticksize = 35
    plt.xticks(size = ticksize)
    plt.yticks(size = ticksize)

    for i in range(len(binedges)-1):

        thebin=np.where((xparameter>binedges[i]) & (xparameter<binedges[i+1]))[0]
        freq,vals=np.histogram(yparameter[thebin],bins=npbins)
        bincenters = [(vals[q]+vals[q+1])/2 for q in range(len(vals)-1)]
        plt.step(bincenters,np.log10(freq+1),linewidth=3,color=colors(i),label=" ".join([str(round(binedges[i],2)),"<",xlabel,"<",str(round(binedges[i+1],2)),";points=",str(len(thebin))]))

    plt.xlabel(ylabel,size=20)
    pylab.legend(loc='upper left',fontsize=15)
    plt.grid("True")
    plt.title(r"log Histogram",size=25)
    
    
    plt.figure (figsize = (20, 10))
    labelsize = 40
    ticksize = 35
    plt.xticks(size = ticksize)
    plt.yticks(size = ticksize)

    for i in range(len(binedges)-1):

        thebin=np.where((xparameter>binedges[i]) & (xparameter<binedges[i+1]))[0]
        freq,vals=np.histogram(yparameter[thebin],bins=npbins)
        bincenters = [(vals[q]+vals[q+1])/2 for q in range(len(vals)-1)]
        plt.step(bincenters,np.log10(freq+1)/np.max(np.log10(freq+1)),linewidth=3,color=colors(i),label=" ".join([str(round(binedges[i],2)),"<",xlabel,"<",str(round(binedges[i+1],2)),";points=",str(len(thebin))]))

    plt.xlabel(ylabel,size=20)
    pylab.legend(loc='upper left',fontsize=15)
    plt.grid("True")
    plt.title(r"rank-ordered log Histogram",size=25)
    
    plt.figure (figsize = (20, 10))
    labelsize = 40
    ticksize = 35
    plt.xticks(size = ticksize)
    plt.yticks(size = ticksize)
    
    mediany=np.median(yparameter)

    for i in range(len(binedges)-1):
        thebin=np.where((xparameter>binedges[i]) & (xparameter<binedges[i+1]))[0]
        freq,vals=np.histogram(yparameter[thebin],bins=npbins)
        bincenters = np.asarray([(vals[q]+vals[q+1])/2 for q in range(len(vals)-1)])
        shiftidx=np.where(np.log10(freq+1)/np.max(np.log10(freq+1))==1)[0]
        theshift=mediany-bincenters[shiftidx]
        plt.step(bincenters+theshift,np.log10(freq+1)/np.max(np.log10(freq+1)),linewidth=3,color=colors(i),label=" ".join([str(round(binedges[i],2)),"<",xlabel,"<",str(round(binedges[i+1],2)),";points=",str(len(thebin))]))

        plt.xlabel(ylabel,size=20)
        pylab.legend(loc='upper left',fontsize=15)
        plt.grid("True")
        plt.title(r"shifted rank-ordered log Histogram",size=25)


    
```
