```python
""""


This cell is used to create a function I call, "exploratoryDA" for exploratory
data analysis. The purpose of this function is to take in data and to perform
a basic exploratory analysis to help the user better understand the story
behind their data. The function takes in exactly 3 arguments: 1.) a numpy array 
which includes the parameters of interest, 2.) a numpy array which includes
the label names of these parameters of interest, and 3.) the number of bins 
the user would like to divide their histograms into. The function takes in
these arguments and then performs three types of analyses:

1.) Histograms: The function outputs one histogram per parameter. The histogram
is divided into the number of bins the user inputs. The histogram includes
a gaussian fit, and to the corner a list of important numbers: the mean, the 
sigma, and the Fisher-Pearson skewness coefficient.

2.) Correlation Matrix: This is a correlation matrix of all of the parameters
inputted into the function. The matrix labels and colors each box by its 
corresponding correlation coefficient (CC).

3.) Scatter Plots: The function makes one scatter plot per each parameter 
combination. The plots are ranked by their CC in descending order. The plots
also include a label with their CC and are colored by this value. 






"""

import seaborn as sn
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import skew
from matplotlib.lines import Line2D


def exploratoryDA(parameters,parameterlabels,nbins):

### MAKING A HISTOGRAM FOR EACH PARAMETER WHICH INCLUDES: GAUSSIAN FIT, MEAN, SIGMA, SKEWNESS
    for i in range(len(parameters)):

        plt.figure (figsize = (10, 10))
        labelsize = 40
        ticksize = 10
        plt.xticks(size = ticksize)
        plt.yticks(size = ticksize)


        plt.hist(parameters[i],bins=nbins, density=True,log=False,fc='forestgreen',ec='black',alpha=0.5)
        mean=np.mean(parameters[i])
        std=np.std(parameters[i])
        skewness=skew(parameters[i])
        xmin, xmax = plt.xlim()
        x = np.linspace(xmin, xmax, 100)
        y = norm.pdf(x, mean, std)
        plt.plot(x, y,linewidth=2,color='black')

        lss = [None,None,None]
        lw= [0,0,0]
        lines = [Line2D([0], [0], color='black', linewidth=lw[l], linestyle=lss[l]) for l in range(len(lss))]
        labels = [" ".join(["mean=",str(round(mean,3))])," ".join(["sigma=",str(round(std,3))])," ".join(["skewness=",str(round(skewness,3))])]
        plt.legend(lines,labels,fontsize=15)


        plt.xlabel(parameterlabels[i],size=20)
        plt.grid("True")
        plt.title("Histogram")

### MAKING A CORRELATION MATRIX FOR ALL PARAMETERS

    data = parameters
    df = pd.DataFrame(np.transpose(data),columns=parameterlabels)
    corr = df.corr()
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    with sn.axes_style("dark"):
        f, ax = plt.subplots(figsize=(20, 10))
        ax = sn.heatmap(corr,annot=True,linewidths=0.5, mask=mask, square=True,cmap='PiYG_r')

    plt.title("Correlation Matrix")


### MAKING A SCATTER PLOT FOR EACH PARAMETER COMBINATION AND RANKING THE PLOT ORDER BY THE CORRELATION COEFFICIENT (CC) DESCENDING 

    df = pd.DataFrame(np.transpose(data))
    corr= df.corr()
    concatcorr=np.unique(np.concatenate(corr))
    rankedcc=np.flip(concatcorr[np.argsort(np.abs(concatcorr))])[1:]
    for i in range(len(rankedcc)):
        paraminds=np.where((corr==rankedcc[i])==True)[0]

        plt.figure (figsize = (10, 10))
        labelsize = 40
        ticksize = 10
        plt.xticks(size = ticksize)
        plt.yticks(size = ticksize)
        
        n=len(parameters[paraminds[0]])
        plt.scatter(parameters[paraminds[0]],parameters[paraminds[1]],c=[rankedcc[i]]*(len(parameters[paraminds[0]])),cmap='PiYG_r')
        plt.xlabel(parameterlabels[paraminds[0]],size=20)
        plt.ylabel(parameterlabels[paraminds[1]],size=20)
        plt.grid("True")
        lss = [None]
        lw= [0]
        lines = [Line2D([0], [0], color='black', linewidth=lw[l], linestyle=lss[l]) for l in range(len(lss))]
        labels = [" ".join(["CC=",str(round(rankedcc[i],3))])]
        plt.legend(lines,labels,fontsize=15)
        cbar=plt.colorbar()
        cbar.set_label(r"CC",size = 20)
        plt.clim(np.min(rankedcc),np.max(rankedcc))
        plt.title("CC-Ranked Scatter Plots")
```
